### Specific instructions:

Your goal here is to practice using astropy constants and units along with python plotting by analyzing blackbody radiation. Recall that blackbody radiation is described by [Planck's law](https://en.wikipedia.org/wiki/Planck%27s_law); the spectral radiance as a function of wavelength is
$$
  B(\lambda,T) = \frac{2hc^2}{\lambda^5}\ \frac{1}{e^{hc/\lambda k T} - 1}
$$
where $\lambda$ is wavelength, $T$ is temperature, $c$ is the speed of light, $h$ is Planck's constant, and $k$ is Boltzmann's constant. The location of the peak of the Planck function is given by [Wien's law](https://en.wikipedia.org/wiki/Wien%27s_displacement_law):
$$
  \lambda_{\rm peak} = \frac{b}{T}
  \qquad\mbox{where}\qquad
  b = 2.897771955 \times 10^{‚àí3} \mbox{ m K}
$$

Your tasks are as follows.

Validation:
- Compute the Planck function for $T = 5000$ K and $\lambda = 0.5\,\mu$m. Print the value (a) in whatever units you get, (b) reduced to m, kg, and s, and (c) in kW sr$^{-1}$ m$^{-2}$ nm$^{-1}$ as on [Wikipedia](https://en.wikipedia.org/wiki/Planck%27s_law). Note that the steradian part (sr$^{-1}$) is implicit, since angles are dimensionless, so you don't actually have to worry about that.
- Use Wien's law to find the peak wavelength for $T = 5000$ K.

Main analysis:
- Plot the Planck function for $T = 3000$ K, 4000 K, and 5000 K. Express the wavelength in $\mu$m and the spectral radiance in kW sr$^{-1}$ m$^{-2}$ nm$^{-1}$, so you can reproduce [this figure on Wikipedia](https://en.wikipedia.org/wiki/Planck%27s_law#/media/File:Black_body.svg). Mark the peak of each curve.

Note: you should not have to specify any [constants](https://docs.astropy.org/en/stable/constants/index.html) or [units](https://docs.astropy.org/en/stable/units/) (including conversions) by hand. Use astropy to handle those.

```python
# your code here

#VALIDATION STARTS HERE

#import all the constants
from astropy import constants as c
from astropy import units as u
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size':16,'text.usetex':True})

#Define the Planck function
def B(wl, temp):return 2*(c.h*c.c**2)*((wl*u.um)**(-5))*(math.exp((c.h*c.c)/(wl*u.um*temp*u.K*c.k_B)) - 1)**(-1)

#whatever units
whatever_units = B(0.5, 5000)
print('Planck function for ùëá = 5000K and ùúÜ = 0.5ùúám in default units is: ', whatever_units)

#reduced to m, kg, and s.
reduced_units = whatever_units.decompose()
print('Planck function for ùëá = 5000K and ùúÜ = 0.5ùúám in reduced units is: ', reduced_units)

#in kW sr^-1, m^-2,nm^-1
weird_units = reduced_units.to(u.kW/u.m**2/u.nm)
print('Planck function for ùëá = 5000K and ùúÜ = 0.5ùúám in Spectral Radiance units is: ', weird_units)

#Use Wien's law to find the peak wavelength at 5000K
def peak_wavelength(temp): return(2.897771955e-3 * u.m)/(temp)
print('The peak wavelength for ùëá = 5000K is: ', peak_wavelength(5000))


#prep
alist = np.linspace(0.01, 3, 1000)
blist, clist, dlist = ([] for i in range(3))
for a in alist: 
    blist.append(B(a, 3000).to(u.kW/u.m**2/u.nm)* u.kW**-1 * u.m * u.m * u.nm)
    clist.append(B(a, 4000).to(u.kW/u.m**2/u.nm)* u.kW**-1 * u.m * u.m * u.nm)
    dlist.append(B(a, 5000).to(u.kW/u.m**2/u.nm)* u.kW**-1 * u.m * u.m * u.nm)

#peak it up
mw = [peak_wavelength(3000)*10**6/u.m, peak_wavelength(4000)*10**6/u.m, peak_wavelength(5000)*10**6/u.m]
temp = [B(mw[0], 3000).to(u.kW/u.m**2/u.nm)* u.kW**-1 * u.m * u.m * u.nm, 
        B(mw[1], 4000).to(u.kW/u.m**2/u.nm)* u.kW**-1 * u.m * u.m * u.nm, 
        B(mw[2], 5000).to(u.kW/u.m**2/u.nm)* u.kW**-1 * u.m * u.m * u.nm]

#plot
plt.xlabel(r'Wavelength ($\mu m)$')
plt.ylabel(r'Spectral Radiance ($kW \cdot sr^{-1} \cdot m^{-2} \cdot nm^{-1} $)', fontsize = 15)
plt.plot(alist, blist,'r', label = r'$3000K$')
plt.plot(alist, clist, 'g', label = r'$4000K$')
plt.plot(alist, dlist, 'b', label = r'$5000K$')
plt.plot(mw, temp, 'x', color = 'k', label = r'$Peak Radiance$')
plt.legend(loc = 'upper right')
plt.show()
```
