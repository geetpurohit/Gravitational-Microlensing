### Specific instructions:
Consider a gravitational lens model that has a singular isothermal sphere with Einstein radius $b$ and an external shear $(\gamma_c,\gamma_s)$ (as described in Section 9.3.5 of my textbook *Principles of Astrophysics* for Physics 341). This lens model is characterized by a lens potential $\phi$ whose first and second derivatives are:
\begin{eqnarray}
  \phi &=& b r + \frac{\gamma_c}{2}(x^2-y^2) + \gamma_s x y \\
  \phi_x &=& \frac{b x}{r} + \gamma_c x + \gamma_s y\\
  \phi_y &=& \frac{b y}{r} + \gamma_s x - \gamma_c y \\
  \phi_{xx} &=& \frac{b y^2}{r^3} + \gamma_c \\
  \phi_{yy} &=& \frac{b x^2}{r^3} - \gamma_c \\
  \phi_{xy} &=& -\frac{b x y}{r^3} + \gamma_s
\end{eqnarray}
where $r = \sqrt{x^2+y^2}$ is the standard polar radius.
* The lens equation relates positions in the image plane $(x,y)$ and source plane $(u,v)$:
\begin{eqnarray}
  u &=& x - \phi_x \\
  v &=& y - \phi_y
\end{eqnarray}
* The lensing magnification is given by $\mu$ such that
$$
  \mu^{-1} = (1-\phi_{xx})(1-\phi_{yy}) - \phi_{xy}^2
$$
* The lensing **critical curve** is the set of points in the image plane where the magnification is infinite, or $\mu^{-1} = 0$.
* The critical curve maps to the **caustic** in the source plane.
* Your goal here is to compute and plot the critical curve and caustic for a few values of the lens model parameters. For this model the critical curve is a closed curve enclosing the origin, so you can define a set of radial rays and use 1-d root finding to find the location of the critical curve on each ray. Consider $b=1$ and the following cases for shear:
  - $(\gamma_c,\gamma_s) = (0.1, 0)$
  - $(\gamma_c,\gamma_s) = (0.2, 0)$
  - $(\gamma_c,\gamma_s) = (0.1, 0.1)$

```python
#%reset -f
# your code here
from astropy import constants as c
from astropy import units as u
from astropy.units import cds
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size':16,'text.usetex':True})
#define θ as theta, γ as gamma, μ as mu

def θ_x(x, y, γ_c, γ_s):
    return x * ((x**2 + y**2)**(0.5))**-1 + (γ_c * x) + (γ_s * y)

def θ_y(x, y, γ_c, γ_s):
    return y * ((x**2 + y**2)**(0.5))**-1 + (γ_s * x) - (γ_c * y) 

def g(x, y, γ_c, γ_s):
    return (x - θ_x(x, y, γ_c, γ_s))

def h(x, y, γ_c, γ_s):
    return (y - θ_y(x, y, γ_c, γ_s))

def θ_xx(x, y, γ_c, γ_s):
    return y**2 * ((x**2 + y**2)**(0.5))**-3 + γ_c

def θ_xx_sub_uv(x, y, γ_c, γ_s):
    return (h(x, y, γ_c, γ_s)**2 * ((g(x, y, γ_c, γ_s)**2 + h(x, y, γ_c, γ_s)**2)**(0.5))**-3 + γ_c)

def θ_yy(x, y, γ_c, γ_s):
    return x**2 * ((x**2 + y**2)**(0.5))**-3 - γ_c

def θ_yy_sub_uv(x, y, γ_c, γ_s):
    return (g(x, y, γ_c, γ_s)**2 * ((g(x, y, γ_c, γ_s)**2 + h(x, y, γ_c, γ_s)**2)**(0.5))**-3 - γ_c)

def θ_xy(x, y, γ_c, γ_s):
    return (-1) * x * y * ((x**2 + y**2)**(0.5))**-3 + γ_s

def θ_xy_sub_uv(x, y, γ_c, γ_s):
    return (-1) * g(x, y, γ_c, γ_s) * h(x, y, γ_c, γ_s) * ((g(x, y, γ_c, γ_s)**2 + h(x, y, γ_c, γ_s)**2)**(0.5))**-3 + γ_s

def μ_inverse(x, y, γ_c, γ_s):
    return ((1 - θ_xx(x, y, γ_c, γ_s))*(1 - θ_yy(x, y, γ_c, γ_s)) - (θ_xy(x, y, γ_c, γ_s))**2)**(-1)

def μ_inverse_uv(x, y, γ_c, γ_s):
    return ((1 - θ_xx_sub_uv(x, y, γ_c, γ_s))*(1 - θ_yy_sub_uv(x, y, γ_c, γ_s)) - (θ_xy_sub_uv(x, y, γ_c, γ_s))**2)**(-1)

def axes(n):
    ax[n].axhline(0, alpha=.2)
    ax[n].axvline(0, alpha=.2)

x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
u = np.linspace(-5, 5, 1000)
v = np.linspace(-5, 5, 1000)
x, y = np.meshgrid(x, y)
u, v = np.meshgrid(u, v)


f,ax=plt.subplots(1,2,figsize=(10,5))
axes(0)
ax[0].contour(x, y, μ_inverse(x, y, 0.1, 0), [0], colors='k')
ax[0].set_title(r'$\mu^{-1} = 0$')
axes(1)
ax[1].contour(u, v, μ_inverse_uv(x, y, 0.1, 0), [0], colors='k')
ax[1].set_title('Source Plane')
f.show()

f,ax=plt.subplots(1,2,figsize=(10,5))
axes(0)
ax[0].contour(x, y, μ_inverse(x, y, 0.2, 0), [0], colors='k')
ax[0].set_title(r'$\mu^{-1} = 0$')
axes(1)
ax[1].contour(u, v, μ_inverse_uv(x, y, 0.2, 0), [0], colors='k')
ax[1].set_title('Source Plane')
f.show()

f,ax=plt.subplots(1,2,figsize=(10,5))
axes(0)
ax[0].contour(x, y, μ_inverse(x, y, 0.1, 0.1), [0], colors='k')
ax[0].set_title(r'$\mu^{-1} = 0$')
axes(1)
ax[1].contour(u, v, μ_inverse_uv(x, y, 0.1, 0.1), [0], colors='k')
ax[1].set_title('Source Plane')
f.show()
```
